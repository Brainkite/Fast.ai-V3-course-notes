```python
from fastai.vision import *
from fastai.callbacks.hooks import*
from fastai.utils.mem import *
```

```python
path = untar_data(URLs.CAMVID)
path.ls()
```

```python
path_lbl = path/'labels'
path_img = path/'images'
```

# Subset classes

# Data

```python
# Recover file names from Folder
fnames = get_image_files(path_img)
fnames[:5]
```

```python
lbl_names = get_image_files(path_lbl)
lbl_names[:5]
```

```python
img_f = fnames[2]
img = open_image(img_f)
img.show(figsize=(7,7))
```

```python
get_y_fn = lambda x: path_lbl/f'{x.stem}_P{x.suffix}'
img_f
```

```python
mask = open_mask(get_y_fn(img_f))
mask.show(figsize=(7,7), alpha=1)
```

```python
src_size = np.array(mask.shape[1:])
src_size, mask.data
```

```python
codes = np.loadtxt(path/'codes.txt', dtype=str)
codes
```

# Dataset

```python
size = src_size//2

free = gpu_mem_get_free_no_cache()
# the max size of bs depends on the available GPU RAM
if free > 8200: bs=8
else: bs=4
print(f"using bs={bs}, have {free}MB of GPU RAM free")
```

```python
# DATA BLOCK API
src = (SegmentationItemList.from_folder(path_img)
       .split_by_fname_file('../valid.txt')  # Separate train_ds from valis_ds with text file
       .label_from_func(get_y_fn, classes=codes)) # use func to recover label mask file
```

```python
data = (src.transform(get_transforms(), size=size, tfm_y=True)
        .databunch(bs=bs)
        .normalize(imagenet_stats))
```

```python
data.show_batch(2, figsize=(10,7))
```

```python
data.show_batch(2, figsize=(10,7), ds_type=DatasetType.Valid)
```

# Model

```python
# Convert array to dictionary with classes as dic keys and indexes as dic val
name2id = {v:k for k,v in enumerate(codes)}
```

```python
void_code = name2id['Void']
void_code
```

```python
def acc_camvid(input, target):
    target = target.squeeze(1)
    mask = target != void_code
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()
```

```python
metrics = acc_camvid
```

```python
wd=1e-2
```

```python
learn = unet_learner(data, models.resnet34, metrics = metrics, wd=wd).to_fp16()
```

```python
lr_find(learn)
learn.recorder.plot()
```

```python
lr=2e-4
```

```python
learn.fit_one_cycle(5, slice(lr), pct_start=.9)
```

```python
learn.recorder.plot_losses()
```

```python
learn.save('stage-1')
```

```python
learn.load('stage-1')
```

```python
learn.show_results(rows=3, figsize=(8,9))
```

```python
learn.unfreeze()
```

```python
lrs= slice(lr/400, lr/4)
```

```python
learn.fit_one_cycle(5, lrs, pct_start=.8)
```

```python
learn.recorder.plot_losses()
```

```python
learn.show_results(rows=3, figsize=(8,9))
```

```python
learn.save('stage-2')
```

```python
learn.load('stage-2')
```
