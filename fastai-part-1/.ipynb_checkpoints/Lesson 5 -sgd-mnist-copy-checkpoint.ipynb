## Single-layer NN trainning

```python
class Mnist_Logistic(nn.Module):
    """
    Cr√©ates neural network class with a single linear layer
    """
    def __init__(self):
        super().__init__() #load nn.Module init
        self.lin = nn.Linear(784, 10, bias=True)
        
    def forward(self, xb):
        return self.lin(xb)
```

```python
model = Mnist_Logistic().cuda()
```

```python
y.shape
```

```python
[p.shape for p in model.parameters()]
```

```python
lr = 1e-2
```

```python
loss_func = nn.CrossEntropyLoss()
```

```python
def update(x,y,lr):
    # model prediction for x
    y_hat = model(x)
    
    # Weight decay
    wd = 1e-5
    w2 = 0.
    for p in model.parameters():
        w2 += (p**2).sum()
    
    # add to regular loss
    loss = loss_func(y_hat, y) + wd*w2
    loss.backward()
    
    # update gradients
    with torch.no_grad():
        for p in model.parameters():
            p.sub_( lr * p.grad)
            p.grad.zero_()
            
    return loss.item()
```

```python
losses1 = [update(x,y,lr) for x,y in data.train_dl]
```

```python
fig = plt.figure(figsize=(14, 8), dpi= 80, facecolor='w', edgecolor='k')

plt.plot(losses1)

leg = ['Mnist_Log,SGD, lr=1e-2']
```

## 2-layers NN trainning

```python
class Mnist_NN(nn.Module):
    def __init__(self):
        super().__init__()
        self.lin1 = nn.Linear(784, 50, bias=True)
        self.lin2 = nn.Linear(50,10, bias=True)
    
    def forward(self, xb):
        x = self.lin1(xb)
        x = F.relu(x)
        return self.lin2(x)
```

### Training with optim.Adam (taking care of Wd)

```python
model = Mnist_NN().cuda()
```

```python
def update(x,y,lr):
    y_hat = model(x)
    
    loss = loss_func(y_hat, y)
    loss.backward()
    
    opt = optim.Adam(model.parameters(), lr)
    
    opt.step()
    opt.zero_grad()
    return loss.item()
    
```

```python
losses2 = [update(x,y,1e-3) for x,y in data.train_dl]
```

```python
def update2(x, y, old, lr, mom): #Homecooked ADAM
    y_hat = model(x)
    
    wd = 1e-8
    w2 = 0.
    for p in model.parameters():
        w2 += (p**2).sum()
    
    loss = loss_func(y_hat, y) + wd*w2
    loss.backward()
    with torch.no_grad():
        for p in model.parameters():
            if torch.equal(old[p][0], torch.zeros(0,device='cuda:0')):
                u = p.grad
                v = torch.ones(p.shape, device='cuda:0')
            else:
                u = p.grad * (1-mom) + old[p][0] * mom
                v = p.grad**2 * (1-mom) + old[p][1] * mom
                
            old[p][0] = u
            old[p][1] = v
            
            p.sub_(lr * u * v**(-.5))
            p.grad.zero_()
            
    return loss.item(), old
```

## Write your own nn.Linear func

```python
from torch.nn.parameter import Parameter
```

```python
class NNlinear(nn.Module):
    """
    Cr√©ate layer of weights and bias
    """
    def __init__(self, in_size, out_size, bias:bool=True):
        super().__init__()
        self.in_size = in_size
        self.out_size = out_size
        
        self.weights = Parameter(torch.ones((out_size, in_size), requires_grad=True))
        if bias:
            self.bias = Parameter(torch.ones((out_size), requires_grad=True))
    
    def forward(self, xb):
        return F.linear(xb, self.weights, self.bias)
        #torch.mm(self.weights, xb.transpose(1,0)) + self.bias 
    
#     def extra_repr(self):
#         return 'in_features={}, out_features={}, bias={}'.format(
#             self.in_features, self.out_features, self.bias is not None)
```

## Write ADAM Optim

```python
class Adam():
    
    def __init__(self, params, mom, lr):
        
        self.params = params
        self.mom = mom
        self.lr = lr
        
        self.old = dict()
        for p in params:
            self.old[p] = 2*[torch.zeros(0)]
    
    def update_grads(self, loss):
        loss.backward()
        with torch.no_grad():
            for p in model.parameters():
                if torch.equal(self.old[p][0], torch.zeros(0)):
                    u = p.grad
                    v = p.grad**2 * (1-self.mom) + p.grad * self.mom
                else:
                    u = p.grad * (1-self.mom) + self.old[p][0] * self.mom
                    v = p.grad**2 * (1-self.mom) + self.old[p][1] * self.mom
                self.old[p][0] = u
                self.old[p][1] = v
                
                p.sub_(self.lr * u / torch.Tensor.sqrt(v))
                p.grad.zero_()
                
                
```
