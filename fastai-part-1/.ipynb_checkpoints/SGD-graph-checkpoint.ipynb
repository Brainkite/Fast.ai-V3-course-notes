```python
import torch
import matplotlib.pyplot as plt
from fastai.basics import *
```

```python
n = 100
```

```python
x = torch.ones(n, 2)
x[:,0].uniform_(-1., 1)
x[:5]
```

```python
w_y = torch.tensor([3., 2]) 
w_y
```

```python
y = x@w_y + torch.rand(n)
```

```python
plt.scatter(x[:,0], y);
```

```python
def mse(y_hat, y): 
    return ((y_hat-y)**2).mean()
```

```python
a = tensor(-1.,1)
```

```python
y_hat = x@a
mse(y_hat, y)
```

```python
plt.scatter(x[:,0],y)
plt.scatter(x[:,0],y_hat);
```

```python
a = tensor(-1.,1)
a = nn.Parameter(a)
a
```

```python
def update():
    y_hat = x@a
    loss = mse(y, y_hat)
    
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad(): # stops grad evaluation during update
        a.sub_(lr * a.grad)
        a.grad.zero_()
```

```python
lr = 1e-1
for t in range(100): 
    update()
    plt.scatter(x[:,0],x@a)
plt.scatter(x[:,0],y)
```

```python
from matplotlib import animation, rc
rc('animation', html='jshtml')
```

```python
a = nn.Parameter(tensor(-1.,1))

fig = plt.figure()
plt.scatter(x[:,0], y, c='orange')
line, = plt.plot(x[:,0], x@a)
plt.close()

def animate(i):
    update()
    line.set_ydata(x@a)
    return line,

animation.FuncAnimation(fig, animate, np.arange(0, 100), interval=20)
```

# 3D plot

```python
from mpl_toolkits import mplot3d
```

```python
def loss_wrt_wgts(w1, w2):
    w = torch.Tensor([w1, w2])
    y_hat = x@w
    return mse(y_hat, y)
loss_wgts = np.vectorize(loss_wrt_wgts)
```

```python
w0_range = np.linspace(-20, 20, 50)
w1_range = np.linspace(-20, 20, 50)
mesh = np.meshgrid(w0_range, w1_range)
loss_mesh = loss_wgts(*mesh)
```

```python
fig = plt.figure(figsize=(12,12))
ax = plt.axes(projection='3d')
ax.plot_surface(*mesh, loss_mesh, cmap='viridis', alpha=0.8)
ax.set_xlabel('w0'); ax.set_ylabel('w1'); ax.set_zlabel('Loss')
ax.view_init(30, 20)
```
